```python
# check the columns in dataset where first column i.e. 'Unnamed: 0' define only serial number 
# which not usefull for prediction of data means it in not correlated with dependent variable
# so we can remove it from dataset

df.drop('Unnamed: 0',axis=1,inplace=True)
# df.drop('fnlwgt',axis=1,inplace=True)
```

```python
# check the types of all columns
df.dtypes
#we observe that there are so many columns which is the type of object 
#so we need to convert its data into numeric form using LabelEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Workclass']=le.fit_transform(df['Workclass'])
df['education']=le.fit_transform(df['education'])
df['marital-status']=le.fit_transform(df['marital-status'])
df['occupation']=le.fit_transform(df['occupation'])
df['relationship']=le.fit_transform(df['relationship'])
df['race']=le.fit_transform(df['race'])
df['sex']=le.fit_transform(df['sex'])
df['native-country']=le.fit_transform(df['native-country'])
# df['Workclass']=le.inverse_transform(df['Workclass'])
# print(df['Workclass'].unique())

# sns.countplot(x=df['Workclass'],hue=df['Income'])
```

```python
df.dtypes
```

```python
# from the dataset the column 'capital-loss' and 'capital-gain' column contain so many value with 0 
# which can affect to calculate the prediction 
# so we can convert it into its mean value

# a=df['capital-loss']==0
# print(a.value_counts(),'\n')

# b=df['capital-gain']==0
# print(b.value_counts(),'\n')

print('so we observe that out of all data approx 90% of data value is 0  \n')
df.head()
```

```python
#convert the data 0 value in capital-loss and capital-gain into its mean value

df['capital-loss'].replace(0,df['capital-loss'].mean(),inplace=True)

df['capital-gain'].replace(0,df['capital-gain'].mean(),inplace=True)

# Note:= we dont need to convert its datatype from int to float 
# it will automatically gets converted into float
```

```python
df.head()
```

```python
# now lets split the dependent and independent features 

x=df.iloc[:,:-1]
y=df.iloc[:,-1]

# x.head()
# y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.3)
lr=LogisticRegression().fit(x_train,y_train)

print(lr.score(x_test,y_test))
y_pred=lr.predict(x_test)

print(y_pred)
# print(y)
#from the above score we observe that our model is gud to prediction 
# its give 80% of accuarcy result 
# now lets find its confusion metrix and accuracy of data
```

```python
from sklearn.metrics import classification_report,confusion_matrix

tn,fp,fn,tp=confusion_matrix(y_test,y_pred).ravel()
print(tp,fn)
print(fp,tn)

# cm=confusion_matrix(y_test,y_pred).ravel()
# cm
# from confusion metrix we observe that there are 52 false positive and 7 false negative are predict wrongly
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score,accuracy_score
```

```python
# where y_test is actual data and y_pred is predicted data

print('accuracy score = ',accuracy_score(y_test,y_pred),'\n')
print('recall score = ',recall_score(y_test,y_pred),'\n')
print('precision score =',precision_score(y_test,y_pred),'\n')
print('f1 score =',f1_score(y_test,y_pred),'\n')
```

```python
# we can directly find this all directly using classification report

print(classification_report(y_test,y_pred))
```

```python
# ROC AUC curve is used for finding the best threshold value of the model
# threshold value will determine the 0 and 1 of person makes over $ 50 or not is correctly predicted or not
from sklearn.metrics import roc_curve
fpr,tpr,threshold=roc_curve(y_test,y_pred)
```

```python
print('fpr =',fpr,'\n')
print('tpr =',tpr,'\n')
print('threshold value =',threshold,'\n')
```

```python
# lets plot the roc curve
plt.plot(fpr,tpr,'-r');
plt.xlabel('False Positive Rate');
plt.ylabel('True Positive Rate');
```

```python
# lets find the auc_roc_score
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
