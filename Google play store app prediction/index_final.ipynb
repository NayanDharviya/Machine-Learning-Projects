Machine learning algorithms can then decide in a better way on how those labels must be operated. It is an important pre-processing step for the structured dataset in supervised learning

```python
LE = preprocessing.LabelEncoder() # object creation of LabelEncoder

# feature_list = ['App','Category','Content Rating','Genres','Type','Current Ver']

feature_list = ['Category','Installs','Content Rating','Genres','Type']

map_data = {}

for col in feature_list:
    data[col] = LE.fit_transform(data[col])
    le_map_data = dict(zip(LE.classes_,LE.transform(LE.classes_)))
    map_data[col] = le_map_data
    
print(map_data)
# data.head()
```

```python
# Last Updated Encoding
# %d --> day ( date in numeric )
# %Y --> Year ( Year in numeric)
# %B --> Month ( Month in words )

# data['Last Updated'] = data['Last Updated'].apply(lambda x : time.mktime(datetime.datetime.strptime(x, '%B %d, %Y').timetuple()))
```

```python
# Strip ( remove ) the $ sign from Price

data['Price'] = data['Price'].apply(lambda x : x.strip('$'))
```

```python
# Remove + sign from Installs and replace ',' --> to ' ' 

# data['Installs'] = data['Installs'].apply(lambda x : x.strip('+').replace(',', ''))
```

```python
# create list which contain data in kb format by searching 'k' values in the string
# find its index value and strip the k 
# convert kb --> to mb by /1024 

k_indices = data['Size'].loc[data['Size'].str.contains('k')].index.tolist()

converter = pd.DataFrame(data.loc[k_indices, 'Size'].apply(lambda x: x.strip('k')).astype(float).apply(lambda x: x / 1024).apply(lambda x: round(x, 3)).astype(str))

data.loc[k_indices,'Size'] = converter
```

```python
# Strip the M from mb data

data['Size'] = data['Size'].apply(lambda x: x.strip('M'))
data[data['Size'] == 'Varies with device'] = 0
data['Size'].astype(float)
```

```python
# change dtype to float

data['Reviews'].astype('float')
data['Installs'].astype('float')
```

# Scaling
StandardScaler is that which will transform your data such that its distribution will have a mean value 0 and standard deviation of 1

```python
# StandardScaler formula ==>  x-mean/SD .......where x = data value, maen --> x-bar, SD --> Standard deviation

from sklearn.preprocessing import StandardScaler # formula

for col in data.drop(['Rating','Android Ver','Last Updated','Current Ver','App'],axis=1):
    ss = StandardScaler()
    data[col] = ss.fit_transform(data[[col]])
    
data.head()
```

    ==> Look all data are in same scale

```python
# Split dependeant and independent feature 

features = ['Category', 'Reviews', 'Size', 'Installs', 'Type', 'Price', 'Content Rating', 'Genres']

# features = ['App','Category', 'Reviews', 'Size', 'Installs', 'Type', 'Price', 'Content Rating', 'Genres', 'Last Updated', 'Current Ver']

x = data[features]  # independent features
y = data['Rating']  # dependent feature
```

```python
from sklearn.model_selection import train_test_split  # used to split the data into training and testing part

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state=10) # 30 % testing data and 70% training data
```

```python
# Function which will predict Testing accuracy and Training accuracy and find the error score ( mean Squared error)

list_dict = []
def model(model_object,name):
    model_object.fit(x_train,y_train)
    list_dict.append({'Algorithm':name,'Training':model_object.score(x_train,y_train),'Testing':model_object.score(x_test,y_test)})
    pred = model_object.predict(x_test)
    print('Mean Absolute Error',':',mean_absolute_error(y_test,pred))
    print('Mean Squared Error',':',mean_squared_error(y_test,pred))
```

```python
# Import All algorithm to check the accuracy of the model
# checks for Best Suitable algorithm which will predict the best testing data
# Here we will use Regression Algorithm because our dependent/ target feature value is continues value not categorical value

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
# model(LinearRegression(),'LinearRegression')
# model(DecisionTreeRegressor(max_depth=5,random_state=42),'DecisionTreeRegressor')
model(RandomForestRegressor(max_depth=5,random_state=42),'RandomForestRegressor')
# model(SVR(kernel='linear'),'Support Vector Regression Linear')
# model(SVR(kernel='poly'),'Support Vector Regression Polynomial')
# model(SVR(kernel='rbf'),'Support Vector Regression Radial Basis Function')
# model(KNeighborsRegressor(n_neighbors=3),'KNeighborsRegressor n = 3')
# model(KNeighborsRegressor(n_neighbors=5),'KNeighborsRegressor n = 5')

pd.DataFrame(list_dict)
```
